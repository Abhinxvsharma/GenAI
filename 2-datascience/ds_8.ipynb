```python
# Matplotlib - library visualization 
# pip install matplotlib 

import matplotlib.pyplot as plt
```

```python
# Bar chart
# Scatter Chart 
# Pie Chart
# Line chart 



# bar chart

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

plt.bar(x=fruits, height=sales, width=0.5) # x axis - categories, y-axis - numerical values
plt.show()
```

```python
import seaborn as sns 
# sns.barplot(x=fruits , y=sales)
sns.barplot(x=sales , y=fruits)
```

```python
plt.bar(x=fruits, height=sales, width=0.6, color=['red','yellow','green','blue']) # x axis - categories, y-axis - numerical values
plt.show()
```

```python
plt.bar(x=fruits, height=sales, width=0.6, color=['red','yellow','green','blue']) # x axis - categories, y-axis - numerical values
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title("Sales vs Fruits")
plt.grid(True)
plt.show()
```

```python
plt.barh(y=fruits, width=[120,150,190,200], height=0.4, color=['red','yellow','green','blue']) # x axis - categories, y-axis - numerical values
plt.xlabel('Sales')
plt.ylabel('Fruits')
plt.title("Sales vs Fruits")
plt.show()
```

```python
# Scatter Plot
# both sides numerical data 
import numpy as np
x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])
```

```python
plt.scatter(x=x,y=y,marker='+',color='green')

plt.show()
```

```python
x1 = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y1= np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])

x2= np.array([13, 51, 17, 32, 80, 51, 20, 67, 14, 91])
y2 = np.array([99, 31, 21, 60,9, 81, 43, 17, 35, 77])

plt.scatter(x1,y1, label='Group1', color='red')
plt.scatter(x2,y2, label='Group2', color='green')


plt.legend()#it will show the label of group1 and group2
plt.grid(True)
plt.show()
```

```python
x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])
l=[10,50,9,16,18,20,22,21,7,19]
color=['red','blue','red','red','blue','blue','red','blue','red','red']
plt.scatter(x=x,y=y,s=l,c=color) # size
plt.grid(True)
```

```python
# pie chart 

country=['India','Pakistan','Bangladesh','Nepal','Spain']
values=[200,180,230,100,120]

plt.pie(x=values, labels=country)

plt.show()
```

```python
plt.pie(x=values, labels=country, autopct="%1.1F%%")
plt.title('Countries Data')
plt.grid(True)
plt.show()
```

```python
# line chart 
data=[12,13,14,18,20,24,27,28]
y=[20,21,22,23,24,25,26,27]
plt.plot(data, y)
plt.grid(True)
plt.show()
```
