```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU ,Dropout,Embedding,Input,SimpleRNN
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.datasets import imdb
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)
x_train
```

```python
x_train=pad_sequences(x_train, maxlen=200)
x_test=pad_sequences(x_test,maxlen=200)
x_train
x_test
```

```python
model=Sequential()
```

```python
#LSTM MODEL

model.add(Input(shape=(200,)))
model.add(Embedding(input_dim=5000, output_dim=32))
model.add(LSTM(32))
# model.add(Dense(32,activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
earlystop=EarlyStopping(monitor='val_loss',patience=2,restore_best_weights=True)
```

```python
history=model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test),callbacks=earlystop)
```

```python
# RNN MODEL

hidden_layers=64
vocab_size=5000


model2=Sequential()
model2.add(Input(shape=(200,)))
model2.add(Embedding(input_dim=vocab_size, output_dim=hidden_layers))
model2.add(SimpleRNN(hidden_layers))
model2.add(Dense(1, activation='sigmoid'))

model2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model2.summary()
```

```python
model2.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test),callbacks=earlystop)
```

```python
model3 = Sequential()

model3.add(Input(shape=(200,)))  # keep same as your data
model3.add(Embedding(input_dim=5000, output_dim=32))
model3.add(GRU(32))
model3.add(Dense(1, activation='sigmoid'))

model3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model3.summary()
```

```python
model3.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test), callbacks=earlystop)
```
