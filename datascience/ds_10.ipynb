```python
# data for funnel chart 
data = dict(
    number=[39, 27.4, 20.6, 11, 2],
    stage=["Website visit", "Downloads", "Potential customers", "Requested price", "invoice sent"])

# for area chart 
# df = px.data.gapminder() 

# for pie chart
# df = px.data.tips()
```

```python
# pip install plotly-express 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(data)
```

```python
df
```

```python
px.funnel(data_frame=df,x='number',y='stage')
```

```python
px.pie(data,names='stage',values='number')
```

```python
# for area chart 
df = px.data.gapminder()
df
```

```python
fig = px.area(df, x="year", y="pop", line_group="country", color='continent')
fig.show()
```

```python
stages = ["Website visit", "Downloads", "Potential customers", "Requested price", "invoice sent"]
df_mtl = pd.DataFrame(dict(number=[39, 27.4, 20.6, 11, 3], stage=stages))
df_mtl
df_mtl['office'] = 'Montreal'
df_mtl

df_toronto = pd.DataFrame(dict(number=[52, 36, 18, 14, 5], stage=stages))
df_toronto['office'] = 'Toronto'
df = pd.concat([df_mtl, df_toronto], axis=0)
df
```

```python
px.funnel(data_frame=df,x='number',y='stage', color='office')
```

```python
# Box plot 
# 25th percentile - 1st quartile - Q1 
# 50th - Median
# 75th - 3rd Quartile  Q3

# IQR - inter quartile range - Q3-Q1 
# Lower Fence: Q1 - (1.5 x IQR)
# Upper Fence: Q3 + (1.5 x IQR)

# box plot - outliers 

p={'age':[19,17,20,23,24,25,22,21,29,43,45,67,78]}
df= pd.DataFrame(p)
df
```

```python
px.box(data_frame=df,x='age')
```

```python
# heatmap 
flights= sns.load_dataset('flights')
flights
```

```python
# pivot table 
data= pd.pivot(data=flights,index='year',columns='month',values='passengers')
```

```python
data
```

```python
sns.heatmap(data,linewidths=0.1, annot=True,fmt='0g',cmap='Blues' )
```

```python
px.density_heatmap(flights, x='month', y='year',z='passengers',text_auto=True)
```
