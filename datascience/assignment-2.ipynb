```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
```

```python
zomato=pd.read_csv('zomato_dataset.csv')
z2=zomato.head(900)
```

```python
zomato.head(100)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=zomato,x='City',width=1,color="green",palette='plasma')
plt.show()
```

```python
sns.scatterplot(data=zomato, x='Delivery Rating' , y='Dining Rating')
```

```python
sns.catplot(data=z2, x='City', y='Votes', palette='plasma')
plt.show()
```

```python
plt.figure(figsize=(20,10))

sns.lineplot(data=z2,x='Restaurant Name', y='Delivery Rating',palette='deep')
plt.tight_layout()
plt.show()
```

```python

px.pie(z2,names='Restaurant Name',values='Dining Rating')
```
