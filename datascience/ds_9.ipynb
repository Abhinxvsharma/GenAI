```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# 

titanic= sns.load_dataset('titanic')  # titanic, tips, flights, iris
titanic
```

```python
titanic.loc[:,'class'].value_counts()
```

```python
sns.countplot(data=titanic, x='class')
```

```python
sns.countplot(data=titanic, x='class', hue='survived')#hue means it will show the count of survived and not survived in each class
```

```python
sns.countplot(data=titanic, x='class', hue='alive', order=['Third','Second','First'],palette='plasma')
#palette is used to change the color of the bars.....#order is used to change the order of the bars......
plt.show()
```

```python

sns.countplot(data=titanic, x='class', hue='alive', order=['Third','Second','First'],hue_order=['yes','no'],width=0.5,fill=True, color='green',palette='plasma')# fill=True by default 
plt.show()
```

```python
c=sns.countplot(data=titanic, x='class', hue='alive', order=['Third','Second','First'],hue_order=['yes','no'],width=0.5,fill=True, color='yellow',palette='plasma') # fill=True by default 
c.set_xlabel('Passenger Class')
c.set_ylabel('Count of Passenger')
c.set_title('Count of Passenger in Classes')
c.set_xticklabels(['4thd','2nd','1st'])
plt.show()
```
