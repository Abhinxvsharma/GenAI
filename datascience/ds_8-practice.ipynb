```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# bar chart

city=["hoshiarpur","ludhiana","amritsar","jalandhar"]
population=[100000,200000,300000,400000]

plt.bar(x=city, height=population, width=.4)
plt.show()
```

```python
import seaborn as sns
sns.barplot(y=city,x=population)
```

```python
plt.bar(x=city, height=population, width=.4, color=['red','yellow','green','blue'])
plt.show()
```

```python
plt.bar(x=city, height=population, width=.4, color=['red','yellow','green','blue'])
plt.xlabel("city")
plt.ylabel("population")
plt.title('city and population')
plt.grid(True)
plt.show()
```

```python
plt.barh(y=city, width=[4,7,9,10],height=0.6 ,color=['red','yellow','green','blue'])
plt.xlabel("population")
plt.ylabel("city")
plt.title('city and population')
plt.show()
```

```python
import numpy as np
x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])
```

```python
# pie chart

subjects=["english","maths","science","social science"]
marks=[30,88,50,99]
plt.pie(x=marks, labels=subjects)
plt.show()
```

```python
plt.pie(x=marks,labels=subjects, autopct="%1.1f%%")
plt.title('student data')
plt.grid(True)
plt.show()
```
