```python
import pandas as pd
```

```python
ser= pd.Series([23,56,46,32,10,100,46,89,91])
ser
```

```python
ser>50 # returns a boolean series
```

```python
ser[ser>50] # returns the values that are greater than 50
```

```python
ser[(ser>30) & (ser<70)] # returns the values that are greater than 30 and less than 70
```

```python
# between
ser[ser.between(32,70,inclusive='both')]  #'both', 'neither', 'left', 'right'
```

```python
ser
```

```python
ser[ser==46]
```

```python
# isin()

ser[ser.isin([46])]
```
