```python
import pandas as pd
data = {
    "text": [
        "I love this product",
        "This is a 23523@# bad product. This is very poor product",
        "Amazing experience. I would love to buy it again",
        "Worst service ever. donot buy it, i am gonna give it -1 star",
        "I am very $#  happy",
        "I am very disappointed   2    "
    ],
    "label": [1, 0, 1, 0, 1, 0]   # 1 = Positive, 0 = Negative
}

df = pd.DataFrame(data)
df
```

```python
# Text Cleaning

import re

def clean_text(text):
    text = text.lower()                  # lowercase
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # remove punctuation & numbers
    text = re.sub(r'\s+', ' ', text)        # remove extra spaces
    return text

df['clean_text'] = df['text'].apply(clean_text)
print(df[['text', 'clean_text']])
```

```python
df
```

```python
# Tokenization 
'''  1. Word Tokenization
     2. Sentence Tokenization
     3. Character tokeniztion
'''

from nltk.tokenize import word_tokenize

df['tokens'] = df['clean_text'].apply(word_tokenize)
print(df['tokens'])
df
```

```python
from nltk.tokenize import sent_tokenize
df['sentence_tokens']=df['clean_text'].apply(sent_tokenize)
```

```python
print(df['sentence_tokens'])
df
```

```python
# charaacter tokenization
a="palwinder"
n=len(a)
l=[]
for i in range(n):
    l.append(a[i])
print(l)
```

```python
# after getting the cleaned text and proper tokens we will perform the stopwords removal
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['no_stopwords'] = df['tokens'].apply(remove_stopwords)
print(df['no_stopwords'])
```

```python
df
```

```python
# stemming 
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

df['stemmed'] = df['no_stopwords'].apply(stemming)
print(df['stemmed'])
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatization(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]

df['lemmatized'] = df['no_stopwords'].apply(lemmatization)
print(df['lemmatized'])
```

```python
df['final_text'] = df['lemmatized'].apply(lambda x: " ".join(x))
print(df['final_text'])
```

```python
df
```

```python
df.drop(columns=['text','clean_text','tokens','sentence_tokens','no_stopwords','stemmed','lemmatized'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X = cv.fit_transform(df['final_text']).toarray()
y = df['label']

print("Vocabulary:", cv.vocabulary_)
print("Feature Matrix:\n", X)
```

```python
#TF-IDF vectorizer === Termed Frequency Inverse document frequency vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X = tfidf.fit_transform(df['final_text']).toarray()
```
