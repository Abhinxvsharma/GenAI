```python
import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
text=['Hello My name is abhinav',
     'I am the student of RBU',
     'I am currently doing training in Genai']

tokenized_data=  [word_tokenize(text.lower()) for text in text]
```

```python
tokenized_data
```

```python
model=Word2Vec(tokenized_data, vector_size=100,window=5,min_count=1)
```

```python
print(model.wv['i'])
```

```python
print(model.wv.similarity('currently','doing'))
```

```python
print(model.wv.most_similar('currently'))
```
