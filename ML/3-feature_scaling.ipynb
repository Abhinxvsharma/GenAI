```python
# feature scaling 
import pandas as pd

data= pd.DataFrame({'marks':[12,13,15,19,23,56,100,150,43,22,179,49,88,86]})
data
```

# X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
# X_scaled = X_std * (max - min) + min

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler((0,2))
```

```python
scaler.fit_transform(data.loc[:,['marks']])
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler= StandardScaler()# mean=0 and std=1
```

```python
std_scaler.fit_transform(data.loc[:,['marks']])
```
