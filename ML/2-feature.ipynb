```python
# In cmd  - pip install scikit-learn -- sklearn
# In notebook - !pip install scikit-learn
import pandas as pd
```

```python
data= pd.DataFrame({'Result':['good','bad','good','bad','excellent','good','good','excellent']})
data
```

```python
data.Result.map({'good':1,'bad':2,'excellent':3})# map is used to replace the values in the column with the values provided in the dictionary
```

```python
# label encoder    # it is used to convert the categorical values into numerical values
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
```

```python
# fit, transform, fit_transform 
encoder.fit(data.Result)# it is used to fit the data to the encoder
```

```python
encoder.transform(data.Result)# it is used to transform the data into numerical values
```

```python
data['Label_result']=encoder.transform(data.Result)
```

```python
data
```

```python
print(data.Result)
```

```python
data['Result']=data['Result'].astype('str')
df_pandas_encoded = pd.get_dummies(data, columns=['Result'])#get_dummies is used to convert the categorical values into numerical values by creating new columns for each category and assigning 1 or 0 to indicate the presence or absence of the category in the original column
df_pandas_encoded
```

```python
data.info()
```

```python
data['Result']=data['Result'].astype('object')
data.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

he=OneHotEncoder(sparse_output=False)#sparse_output output ko false karne se hume ek dense array milega jisme 1 aur 0 honge

categorical_columns=data.select_dtypes(include=['object','category']).columns
categorical_columns
```

```python
encoded=he.fit_transform(data[categorical_columns])

```

```python
he.get_feature_names_out(categorical_columns)
```

```python
print(encoded)
```

```python

one_hot_df = pd.DataFrame(encoded, 
                          columns=he.get_feature_names_out(categorical_columns))
```

```python
one_hot_df
```

```python
pd.concat([data,one_hot_df],axis=1)
```
