```python
countries = ["Iran", "Israel", "United States"]

filtered = df[df["Country"].isin(countries)]
filtered
```

```python
fig = px.scatter(
    filtered,
    x="Country",
    y="Per 1,000 capita(total)",
    size="Total",
    title="Military Personnel per 1000 Capita (Iran, Israel, USA)",
    hover_name="Country"
)

fig.show()
```

```python
# ONE HOT ENCODING
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df=pd.DataFrame({"color":['red','blue','green','red']})

one= OneHotEncoder(sparse_output=False)

ones=one.fit_transform(df.loc[:,["color"]])

ones_df=pd.DataFrame(ones,columns=one.get_feature_names_out())

ones_df
```

```python
pd.concat([df,ones_df],axis=1)
```
