```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df
```

```python
df['target']=data.target
```

```python
df
```

```python
x=df.drop('target',axis=1)
```

```python
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_train_sc
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=8)
```

```python
knn
```

```python
knn.fit(x_train_sc,y_train)
```

```python
knn.score(x_test_sc,y_test)
```

```python
y_pred=knn.predict(x_test_sc)
y_pred
```

```python
np.array(y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cf=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
```

```python
sns.heatmap(cf,annot=True,cmap='Greens')
plt.xlabel('Actual Values')
plt.ylabel('Predicted values')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Import Libraries
import pandas as pd
import plotly.express as px

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensions to 2 using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame
df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df["target"] = y

# Map class names
df["diagnosis"] = df["target"].map({0: "Malignant", 1: "Benign"})

# Plot interactive scatter plot
fig = px.scatter(
    df,
    x="PC1",
    y="PC2",
    color="diagnosis",
    title="Breast Cancer Dataset Cluster Visualization (PCA)",
    labels={"PC1": "Principal Component 1", "PC2": "Principal Component 2"}
)

fig.show()
```

```python
pca_components=pd.DataFrame(pca.components_,columns=data.feature_names,index=['PCA1','PCA2'])
pca_components
```
